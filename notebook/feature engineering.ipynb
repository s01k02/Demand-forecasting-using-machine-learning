step2 sort data by stores and date


```python

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(["Store", "Date"])

```

```python
df.head()
```

CREATE LAG FEATURES( stores about the last sales )

```python
df["lag_1"] = df.groupby("Store")["Weekly_Sales"].shift(1)
df["lag_2"]=df.groupby("Store")["Weekly_Sales"].shift(2)
df["lag_4"]=df.groupby("Store")["Weekly_Sales"].shift(4)
```

 STEP3 calculate the ROLLING MEANS (BEACAUSE  smooth noise and captures trends)

```python
df["rolling_mean_4"] = (
    df.groupby("Store")["Weekly_Sales"]
      .rolling(4)
      .mean()
      .reset_index(level=0, drop=True)
)
```

```python
df=df.dropna()
```
